## 1.1
Definitions: 

- $ \|x||_{\infty} = \underset{i}{max} |x_i| $

- $ \|x\|_1 = (|x_1| + |x_2| + ... + |x_n|) $

Let $ x_k = \underset{i}{max} |x_i| $

<br>

$ \|x\|_{\infty} \leq C_2 \| x\|_{1} $

$ |x_k| \leq C_2(|x_1| + |x_2| +...+|x_k|+... + |x_n|) \iff |x_k| \leq C_2|x_k| + C_2(|x_1| + |x_2| +...+ |x_n|) $

$ \forall n: |x_1| + |x_2| +...+ |x_n| \geq 0 $

That means $ C_2 = 1: |x_k| \leq |x_k| + (|x_1| + |x_2| +...+ |x_n|) \implies \|x\|_{\infty} \leq C_2 \| x\|_{1}, C_2 = 1 $

<br>

$C_1 \|x\|_{1} \leq \|x\|_{\infty} $

$ C_1(|x_1| + |x_2| +...+|x_k|+... + |x_n|) \leq |x_k|\overset{divider\geq0}{\iff} C_1 \leq \frac{|x_k|}{|x_1|+|x_2|+...+|x_k|+... + |x_n|} \leq \frac{|x_k|}{|x_k|+|x_k|+...+|x_k|+...+|x_k|} = \frac{1}{n} $

<br>

$$ C_1 = \frac{1}{n}, C_2 = 1 $$

## 1.3

### Computation Complexity and Automatic Differentiation with JAX

Because of symmetry, complexity of computing gradient with respect to U and to V will be the same. $ \frac{\partial F}{\partial V} = DU^T(UDV-X)$

$X \in \mathbb{R}^{n \times n}$, $U \in \mathbb{R}^{n \times k}$, $V \in \mathbb{R}^{k \times n}$, $k < n$ and $D = diag(d_1, \ldots, d_k)$ is given diagonal matrix.

To compute $DV$ we need 1 multiplication for entry, thus, $O(kn)$

To compute $U_{nxk}(DV)_{kxn}$m, thus, $O(kn^2)$

Then $nxn$ subtractions in $UDV - X$, $O(n^2)$.

In $(U^*)_{kxn}(UDV - X)_{nxn}$ that means $O(kn^2)$

The last step $DU^*(UDV-X)$ as a first step takes $O(kn)$

Finally, $ O(kn + kn^2 + n^2 + kn^2 + kn) \implies O(kn^2)$

The compliesity is: $O(kn^2)$

````python
# Define functions for gradient calculation

import jax.numpy as jnp
import jax
import numpy as np
from numba import jit
import timeit

# Naive calculation
def deriv_V(U, V, D, X):
    analytical_value = D @ U.T @ (U @ D @ V - X)
    return analytical_value

def deriv_U(U, V, D, X):
    analytical_value = (U @ D @ V - X) @ V.T @ D 
    return analytical_value

# Automatic differentiation with JAX
@jax.jit
def auto_dif(U, V, D, X):
    f = jnp.square((jnp.linalg.norm((X - U @ D @ V), 'fro'))) / 2
    return f
````

````python
# In this loop, I set k and vary n. Results of calculations of gradient are saved in lists with index 'n'.

U_exp_n = []
V_exp_n = []
U_jax_n = []
V_jax_n = []

n = 200
k = 150

for i in range(200, 1000, 25):
    n = i
    diag = jax.random.normal(jax.random.PRNGKey(0), (1, k)) #Diagonal entries for D
    D = np.zeros((k, k))
    np.fill_diagonal(D, diag)

    X = jax.random.normal(jax.random.PRNGKey(0), (n, n)) #Random n x n matrix
    U = jax.random.normal(jax.random.PRNGKey(0), (n, k)) #Random n x k matrix
    V = jax.random.normal(jax.random.PRNGKey(0), (k, n)) #Random k x n matrix

    der_V = timeit.timeit(lambda: deriv_V(U, V, D, X), number=8)
    der_U = timeit.timeit(lambda: deriv_U(U, V, D, X), number=8)
    
    V_jax = jax.jit(jax.grad(auto_dif, argnums=1))
    t_V_jax = timeit.timeit(lambda: V_jax(U, V, D, X), number=8)

    U_jax = jax.jit(jax.grad(auto_dif, argnums=0))
    t_U_jax = timeit.timeit(lambda: U_jax(U, V, D, X), number=8)
    
    U_exp_n.append(der_U * 1000)
    V_exp_n.append(der_V * 1000)
    U_jax_n.append(t_U_jax * 1000)
    V_jax_n.append(t_V_jax * 1000)
````

````python
# In this loop, I set n and vary k. Results of calculations of gradient are saved in lists with index 'k'.

U_exp_k = []
V_exp_k = []
U_jax_k = []
V_jax_k = []

n = 1000
k = 150

for i in range(150, 1000, 25):
    k = i
    diag = jax.random.normal(jax.random.PRNGKey(0), (1, k)) #Diagonal entries for D
    D = np.zeros((k, k))
    np.fill_diagonal(D, diag)

    X = jax.random.normal(jax.random.PRNGKey(0), (n, n)) #Random n x n matrix
    U = jax.random.normal(jax.random.PRNGKey(0), (n, k)) #Random n x k matrix
    V = jax.random.normal(jax.random.PRNGKey(0), (k, n)) #Random k x n matrix

    der_V = timeit.timeit(lambda: deriv_V(U, V, D, X), number=8)
    der_U = timeit.timeit(lambda: deriv_U(U, V, D, X), number=8)
    
    V_jax = jax.jit(jax.grad(auto_dif, argnums=1))
    t_V_jax = timeit.timeit(lambda: V_jax(U, V, D, X), number=8)

    U_jax = jax.jit(jax.grad(auto_dif, argnums=0))
    t_U_jax = timeit.timeit(lambda: U_jax(U, V, D, X), number=8)
    
    U_exp_k.append(der_U * 1000)
    V_exp_k.append(der_V * 1000)
    U_jax_k.append(t_U_jax * 1000)
    V_jax_k.append(t_V_jax * 1000)
````

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class ResidualBlock(nn.Module):
    def __init__(self, in_features, middle_features, activation=nn.ReLU()):
        super().__init__()
        self.layer1 = nn.Linear(in_features, middle_features)
        self.layer2 = nn.Linear(middle_features, in_features)
        self.activation = activation
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.activation(out)
        out = self.layer2(out)
        out = out + x
        out = self.activation(out)
        return out

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.res1 = ResidualBlock(1000,1000)
        self.res2 = ResidualBlock(1000,1000)
        self.fc2 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.train()
    model.to(device)
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward(retain_graph=True)
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
````

#### 3.2. Creating a new net

````python
def new_layer(layer, r):
    A = layer.weight.data
    u, s, v = torch.linalg.svd(A)
    u = u[:, :r]
    s = torch.diag(s[:r])
    v = v[:r, :]
    b = layer.bias.data
    return u, s, v, b
````

````python
class new_ResidualBlock(nn.Module):
    def __init__(self, model_res, r, activation=nn.ReLU()):
        super().__init__()
        u1, s1, v1, b1 = new_layer(model_res.layer1, r)
        u2, s2, v2, b2 = new_layer(model_res.layer2, r)
                
        self.u1 = nn.Parameter(u1, requires_grad=True)
        self.s1 = nn.Parameter(s1, requires_grad=True)
        self.v1 = nn.Parameter(v1, requires_grad=True)
        self.b1 = nn.Parameter(b1, requires_grad=True)
        
        self.u2 = nn.Parameter(u2, requires_grad=True)
        self.s2 = nn.Parameter(s2, requires_grad=True)
        self.v2 = nn.Parameter(v2, requires_grad=True)
        self.b2 = nn.Parameter(b2, requires_grad=True)
        
        self.activation = activation
        
    def forward(self, x):
        out = x @ (self.u1 @ self.s1 @ self.v1).T + self.b1
        out = self.activation(out)
        out = out @ (self.u2 @ self.s2 @ self.v2).T + self.b2
        out = out + x
        out = self.activation(out)
        return out

class new_Net(nn.Module):
    def __init__(self, model, r):
        super(new_Net, self).__init__()
        self.fc1 = model.fc1
        self.res1 = new_ResidualBlock(model.res1, r)
        self.res2 = new_ResidualBlock(model.res2, r)
        self.fc2 = model.fc2
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````

````python
# Updated test function
# The only difference with the origin is that I return an accucary in the end

def new_test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
    
    acc_perc = 100. * correct / len(test_loader.dataset)
    return acc_perc
````

## Problem 4. «Reducio!» (15 pts)

````python
import jax.numpy as jnp
import jax
import numpy as np
import random
import math

class Game:
    def __init__(self, key):
        # key is a jax.random.PRNGKey
        self.key = key
        return
    def hermione_chooses_r_and_W(self):
        self.r = random.randint(1, 95)
        self.key, subkey = jax.random.split(self.key)
        self.W1 = jax.random.uniform(subkey, (self.r, 100), maxval=100., minval=0., dtype=jnp.float32)
        self.key, subkey = jax.random.split(self.key)
        self.W2 = jax.random.uniform(subkey, (100, self.r), maxval=100., minval=0., dtype=jnp.float32)
    def hermione_computes_function(self, x):
        return jnp.sum(jnp.square(jnp.sin(self.W2 @ jnp.cos(self.W1 @ x))))
    def harry_answers(self):
        # <your code here>
        # you shouldn't use self.r, self.W1, or self.W2
        # you can call `hermione_computes_function` multiple times

        harrys_vector = jax.random.normal(jax.random.PRNGKey(0), (1, 100))
        #print(harrys_vector.shape)

        for i in range(1, 100):
            i = jax.random.normal(jax.random.PRNGKey(0), (1, 100))
            harrys_vector = np.append(harrys_vector, i, axis=0)
        #print(harrys_vector.shape)

        snapes_grad = jax.jit(jax.jacfwd(self.hermione_computes_function))
        rons_matrix = snapes_grad(harrys_vector)
        #print(rons_matrix.shape)
        
        rons_matrix = rons_matrix / jnp.sqrt(harrys_vector.shape[0])
        _, dobby, _= jnp.linalg.svd(rons_matrix)
        
        return dobby[dobby > 1].shape[0]


    def play(self, n_rounds, verbose=True):
        # n_rounds: a number of rounds of the game
        # verbose: print or not the result of each round
        n_right_answers = 0
        for _ in range(n_rounds):
            self.hermione_chooses_r_and_W()
            r = self.harry_answers()
            if abs(r - self.r) == 0:
                if verbose:
                    print("Good job! The true answer is {}, Harry's answer is {}!".format(self.r, r))
                n_right_answers += 1
            else:
                if verbose:
                    print("Harry's answers is {}, but the true answer is {} :(".format(r, self.r))
        if float(n_right_answers) / n_rounds > 0.95:
            print('Well done: {}/{} right answers!'.format(n_right_answers, n_rounds))
        else:
            print('Only {}/{} right answers :( Work a little more and you will succeed!'.format(
                n_right_answers, n_rounds))
````
