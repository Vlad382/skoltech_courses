## Experiment with a search ellipsoid

```python
from generator import *
input_folder = 'D:/spe10_mod2/input_folder/'
output_folder = 'C:/Users/Vlad/Desktop/geostat_pr/'

gen = Generator(root=input_folder, path_to_save=output_folder, model='spe10m2', alg='ord_kriging')
```

```python
size, number = 1, 1
alpha = .3
prop_name = 'poro'
gen.search_ellipsoid = '450 300 100  0 90 0'
```

```python
gen.run_name = 'choose_ellip_1'
gen._set_generation_mode(size, alpha)
gen._upload_data()        
gen._match_coord()

gen.work_dir = abs_join_path(gen.size_dir, str(alpha*100)+'%_'+str(number))
gen._take_points()


os.mkdir(gen.work_dir)
print('Saving GSLIB files to: ' + gen.work_dir)
gen._save_gslib_all(number)

gen.est_results = {}
gen.property_name = prop_name
gen._set_algorithm()
```

```python
# Can change parameters of the chosen algorithm
gen._print_alg_tree()
```

```python
search_ellipsoid = '800 700 200  0 90 0'
```

```python
gen._change_alg_params(search_ellipsoid)
```

```python
gen._run_alg()
kriging_res = pd.read_csv(abs_join_path(gen.path_to_save, 'results/results.grid'), skiprows=3, names=[gen.property_name]).to_numpy()
kriging_res = np.clip(kriging_res, a_min=0, a_max=None).ravel()
gen.est_results[gen.property_name] = kriging_res
print(f'For {gen.property_name} in run {number} estimation is completed.')

num_nan = np.where(kriging_res == -9966699)[0].shape[0]
if num_nan:
    num_of_points = gen.dimens[0] * gen.dimens[1] * gen.dimens[2]
    print(f"'Warning!\n'In run {number} property {gen.property_name} has {np.round(num_nan*100/num_of_points, 2)}% of NaN's in the results.\n")
```

```python
gen.PlotResults(p_name='poro')
```
